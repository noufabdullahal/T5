### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(10, 6))
plt.hist(df['TrafficIndexLive'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution ')
plt.xlabel('Traffic_lndex')
plt.ylabel('weather_Condition')
plt.show()
```

## 3. Insights and Conclusions

```python

plt.figure(figsize=(10, 6))
city_traffic_index = df.groupby('City')['TrafficIndexLive'].sum()
city_traffic_index.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Traffic Index Live Distribution by City')
plt.ylabel('')
plt.show()
```
